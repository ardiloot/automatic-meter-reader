```python
import cv2
import matplotlib.pyplot as plt
from automatic_meter_reader import AutomaticMeterReader
```

## Read image

Must be in BGR format (default in OpenCV)

```python
img = cv2.imread("tests/test_image.jpg")
```

## Automatic readout

```python
camera_model = "espcam_120_deg"
meter_model = "wehrle_hot"
amr = AutomaticMeterReader(camera_model, meter_model)
measurement = amr.readout(img)

if measurement is not None:
    print("Measurement: %.3f" % (measurement))
else:
    print("Automatic readout failed!")
```

# Visualize

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

axs[0].set_title("Original image")
axs[0].imshow(cv2.cvtColor(amr.img_original, cv2.COLOR_BGR2RGB))

axs[1].set_title("Undistorted and aligned image")
axs[1].imshow(cv2.cvtColor(amr.img_debug, cv2.COLOR_BGR2RGB))

for ax in axs:
    ax.axis("off")
```
